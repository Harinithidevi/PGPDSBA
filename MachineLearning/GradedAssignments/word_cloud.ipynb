```python
from nltk.corpus import stopwords
from nltk.tokenize import RegexpTokenizer
from collections import Counter
```

```python
tokenizer=RegexpTokenizer(r'\w+')
```

```python
roosevelt_no_punc=tokenizer.tokenize(Roosvelt)
set(w.title() for w in roosevelt_no_punc if w.lower() not in stopwords)
word_count_punc = Counter(w.title() for w in roosevelt_no_punc if w.lower() not in stopwords)
word_count_punc.most_common()
```

```python
roosevelt_no_punc=tokenizer.tokenize(Kennedy)
set(w.title() for w in roosevelt_no_punc if w.lower() not in stopwords.words())
word_count_punc = Counter(w.title() for w in roosevelt_no_punc if w.lower() not in stopwords.words())
word_count_punc.most_common()
```
