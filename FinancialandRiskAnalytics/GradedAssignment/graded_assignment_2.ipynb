```python
stock_returns=np.log(stock_prices.drop(['Date', 'dates'], axis=1)).diff(axis=0, periods=1)
```

```python
stock_returns.shape
```

```python
stock_returns.head()
```

```python
stock_ret_sum = stock_returns.sum(axis=0)
stock_ret_sum
```

```python
stock_means = stock_returns.mean(axis=0)
stock_means
```

```python
stock_means_sd=stock_returns.std(axis=0)
```

```python
df=pd.DataFrame({'Average':stock_means, 'Volatility':stock_means_sd})
```
